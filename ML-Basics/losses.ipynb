```python
x = symbols("x")

logloss = -log(x)
diff_logloss = diff(logloss)

plot(logloss, (x, 1e-2, 1))
plot(diff_logloss, (x, 1e-2, 1))
```

# Pinball / Quantile Loss


```python
# @np.vectorize
def quantile_loss(q, y, yhat):
    if y > yhat:
        return (y - yhat) * q
    else:
        return (yhat - y) * (1 - q)
```

```python
x = np.arange(-10, 10, 0.1)
y_95 = [quantile_loss(0.95, 0, elem) for elem in x]
y_05 = [quantile_loss(0.05, 0, elem) for elem in x]
plt.plot(x, y_95, label="95")
plt.plot(x, y_05, label="05")
plt.legend()
```
